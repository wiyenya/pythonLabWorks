# Памʼятки Культури

### Памʼятки Культури

- **Гейсіра:**
  - Геотермальна зона, розташована в південній частині Ісландії.
  - Один з найвідоміших гейзерів в зоні – Строккур:
    - Вивергається приблизно кожні 5-10 хвилин.
    - Може досягати висоти до 20 метрів.
  - Інші гейзери в зоні:
    - Гейзер Гейсіра (звідки назва зони) практично не вивергається, але є історично важливим.

- **Рейк’явік:**
  - Столиця Ісландії, розташована на південно-західному узбережжі острова.
  - Відомий сучасний архітектурний дизайн:
    - Будівля Харпа – концертний зал і конференц-центр з унікальною фасадною конструкцією.
    - Халльгрімскіркья – церква з сучасним дизайном і панорамним видом на місто з вежі.
  - Культурні події та активне нічне життя:
    - Щорічний музичний фестиваль Iceland Airwaves.
    - Різноманітні музеї, галереї та театри.
    - Популярні ресторани і бари в районі центру міста.

- **Сага-музей:**
  - Музей, розташований у Рейк’явіку.
  - Представляє історію Ісландії та її саґи:
    - Експозиції, що відображають важливі події з історії та літературної спадщини.
    - Інтерактивні виставки, що дозволяють зануритися в середньовічну ісландську культуру.
  - Спеціальні заходи та виставки:
    - Музей організовує спеціальні заходи, лекції та тематичні виставки.

```python
def decimal_to_binary(n): 
    if n == 0:
        return "0"
    binary = ""
    while n > 0:
        remainder = n % 2
        binary = str(remainder) + binary
        n = n // 2
    return binary
    

def fractional_to_binary(fraction, precision=10): 
    if not (0 <= fraction < 1):
        return "Fraction must be between 0 and 1"
    
    binary = ""
    while precision > 0 and fraction != 0:
        fraction *= 2
        bit = int(fraction)
        if bit == 1:
            binary += "1"
            fraction -= bit  # Уменьшаем на 1, чтобы продолжить работу с дробной частью
        else:
            binary += "0"
        precision -= 1
        
    return binary


def task_7(num):
    result_str = ''
    
    if num>0:
        result_str = '0|' + result_str
    else:
        result_str = '1|' + result_str
    
    num_str = str(abs(num))
    nums_arr = []
    
    if "." in num_str:
        nums_arr = num_str.split(".")
        nums_arr[1] = '0.' + nums_arr[1]
    else:
        nums_arr.append(num_str)
    
    
    num1 = decimal_to_binary(int(nums_arr[0]))
    if len(nums_arr) > 1:
        num2 = fractional_to_binary(float(nums_arr[1]))
    
    
    if len(nums_arr) > 1:
        result_str = result_str + num1 + '.' + num2
    else :
        result_str = result_str + num1

    print (result_str)
    
    
    
    
```
